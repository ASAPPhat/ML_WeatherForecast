```python
from datetime import datetime, timedelta
import os
import json 
start_date = datetime(2024,8,11)
end_date = datetime(2024,9,11)
current= start_date
while current < end_date: 
    date_str = current.strftime('%Y-%m-%d')
    rsp = requests.get(f'http://api.weatherapi.com/v1/history.json?key=4a99b1361157474eb9730458241209&q="Ho Chi Minh"&dt={date_str}').json()
    current+= timedelta(days=1)
    with open(f"weather_{date_str}.json", "w") as f:
        json.dump(rsp, f, indent=4)
    name = rsp['location']['name']
    country = rsp['location']['country']
    localtime = rsp['location']['localtime']
    print(f'Date: {date_str}, City: {name}, Country: {country}, Local Time: {localtime}')
```

```python
import requests
import pandas as pd
from datetime import datetime, timedelta

def get_weather_data(api_key, city, date):
    base_url = f'http://api.weatherapi.com/v1/history.json?key={api_key}&q={city}&dt={date}'
    response = requests.get(base_url).json()
    return response

def date_range(start_date, end_date):
    current_date = start_date
    while current_date <= end_date:
        yield current_date.strftime('%Y-%m-%d')
        current_date += timedelta(days=1)

def collect_weather_data(api_key, city, start_date, end_date):
    weather_records = []
    for date in date_range(start_date, end_date):
        data = get_weather_data(api_key, city, date)
        try:
            weather_info = {
                'city': city,
                'date': date,
                'temperature': data['forecast']['forecastday'][0]['day']['avgtemp_c'],
                'humidity': data['forecast']['forecastday'][0]['day']['avghumidity'],
                'pressure': data['forecast']['forecastday'][0]['day'].get('pressure_mb', 'Unknown'),
                'weather': data['forecast']['forecastday'][0]['day']['condition']['text'],
                'wind_speed': data['forecast']['forecastday'][0]['day']['maxwind_kph'],
                'timestamp': datetime.now()
            }
            weather_records.append(weather_info)
        except KeyError as e:
            print(f"KeyError for date {date}: {e}")
            continue
    
    return weather_records

# Define API key, city, and date range
api_key = "4a99b1361157474eb9730458241209"
city = "Ho Chi Minh"
start_date = datetime.strptime('2024-02-01', '%Y-%m-%d')
end_date = datetime.strptime('2024-04-02', '%Y-%m-%d')

# Collect weather data
weather_data = collect_weather_data(api_key, city, start_date, end_date)

# Create a DataFrame
df = pd.DataFrame(weather_data)

# Save DataFrame to JSON
df.to_json('weather_data.json', orient='records', lines=True)

print("Data saved to weather_data.json")
```
